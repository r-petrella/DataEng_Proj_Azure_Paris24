## CREATE THE PARAMETERS WITH A DEFAULT VALUE AND THEN RETRIEVE THE DYNAMIC VALUES FOR EACH PARAMETER FROM THE PARAMETERS WITHIN THE JOB TASK

#### Here we do not make data transformation. We just fetch the data from bronze and we store them in silver changing the format to delta

source_catalog = "olympicspariscatalog"
folder = "athletes"
sink_container = "gold"

```python
# create the initial parameter as an empty string
dbutils.widgets.text("source_catalog", "")
# retrieve the parameter value from the job (the right value is extracted through the job parameters)
source_catalog = dbutils.widgets.get("source_catalog")

# create the initial parameter as an empty string
dbutils.widgets.text("sink_container", "")
# retrieve the parameter value from the job (the right value is extracted through the job parameters)
sink_container = dbutils.widgets.get("sink_container")

# create the initial parameter as an empty string
dbutils.widgets.text("folder", "")
# retrieve the parameter value from the job (the right value is extracted through the job parameters)
folder = dbutils.widgets.get("folder")
```

##IMPORT THE TABLES FROM CATALOG.GOLD

```python
%python
df = spark.table(f"{source_catalog}.{sink_container}.{folder}")
```

## WRITE THE TABLES TO SILVER

```python
df.write.format("parquet")\
  .mode("overwrite")\
  .option("path", f"abfss://{sink_container}@olympicsprogectstorage.dfs.core.windows.net/{folder}")\
  .save()
```
